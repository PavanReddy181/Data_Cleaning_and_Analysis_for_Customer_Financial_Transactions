```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the Excel sheets
df_acquisition = pd.read_csv('customer acquisition.csv')
df_spend = pd.read_csv('Spend.csv')
df_repayment = pd.read_csv('Repayment.csv')
```

```python
df_acquisition
```

```python
df_spend.rename(columns={'Costomer':'Customer'},inplace=True)
df_spend
```

```python
customer_spending = df_spend.groupby('Customer')['Amount'].sum().reset_index()
customer_spending
```

```python
df_repayment.rename(columns={'Costomer':'Customer'},inplace=True)
df_repayment.rename(columns={'Amount':'Repaid'},inplace=True)
df_repayment
```

# Task a: Provide a meaningful treatment to values where age is less than 18

```python
df_acquisition['Age'] = pd.to_numeric(df_acquisition['Age'], errors='coerce')
```

```python
mean_age = np.mean(df_acquisition.loc[df_acquisition['Age'] >= 18, 'Age'])
mean_age
```

```python
df_acquisition.loc[df_acquisition['Age'] < 18, 'Age'] = mean_age
```

```python
df_acquisition
```

# Task b: Identify where repayment is more than spend and give them a credit of 2% of surplus amount in next month billing

```python
merged_df = pd.merge(df_spend, df_repayment, on=['Customer', 'Month'], suffixes=('_spend', '_repayment'))
merged_df['Surplus'] = merged_df['Repaid'] - merged_df['Amount']
customer_surplus = merged_df[['Customer', 'Surplus']]
customer_surplus 
```

```python
credit_increase = merged_df.loc[customer_surplus['Surplus'] > 0, ['Customer', 'Surplus']]
credit_increase = credit_increase.groupby('Customer')['Surplus'].sum().reset_index()
credit_increase['Credit Increase'] = credit_increase['Surplus'] * 0.02
credit_increase = credit_increase[['Customer', 'Credit Increase']]
credit_increase
```

# Task 1: Monthly spend of each customer

```python
monthly_spend = df_spend.groupby('Customer')['Amount'].sum().reset_index()
```

```python
monthly_spend = monthly_spend.sort_values('Customer', ascending=True)
monthly_spend 
```

# Task 2: Monthly repayment of each customer

```python
monthly_repayment = df_repayment.groupby('Customer')['Repaid'].sum().reset_index()
```

```python
monthly_repayment
```

# Task 3: Highest paying 10 customers

```python
highest_paying_customers = df_repayment.groupby('Customer')['Repaid'].sum().nlargest(10).reset_index()
highest_paying_customers
```

# Task 4: People in which segment are spending more money

```python
segment_spending = pd.merge(df_acquisition, df_spend, on='Customer')
segment_spending = segment_spending.groupby('Segment')['Amount'].sum().reset_index()
segment_spending = segment_spending.sort_values('Amount', ascending=False)
segment_spending = segment_spending.reset_index(drop=True)
segment_spending 
```

# Task 5: Age group wise money spending

```python
age_groups = pd.cut(df_acquisition['Age'], bins=[0, 18, 25, 35, 50, float('inf')], 
                    labels=['Under 18', '18-25', '26-35', '36-50', '50+'])

age_group_spending = pd.merge(df_spend, df_acquisition.assign(Age_Group=age_groups), on='Customer')
age_group_spending = age_group_spending.groupby('Age_Group')['Amount'].sum().reset_index()
age_group_spending
```

# Task 6: Determine the most profitable segment

```python
most_profitable_segment = pd.merge(df_acquisition, df_spend, on='Customer')
most_profitable_segment = pd.merge(most_profitable_segment, df_repayment, on='Customer')
most_profitable_segment['Repaid'] = pd.to_numeric(most_profitable_segment['Repaid'], errors='coerce')
most_profitable_segment['Amount'] = pd.to_numeric(most_profitable_segment['Amount'], errors='coerce')
most_profitable_segment['Profit'] = most_profitable_segment['Repaid'] - most_profitable_segment['Amount']
most_profitable_segment = most_profitable_segment.groupby('Segment')['Profit'].sum().idxmax()
most_profitable_segment = pd.DataFrame({'Most Profitable Segment': [most_profitable_segment]})
most_profitable_segment
```

# Task 7: In which category are customers spending more money

```python
category_spending = pd.merge(df_acquisition, df_spend, on='Customer')
category_spending = category_spending.groupby('Type')['Amount'].sum().reset_index()
category_spending = category_spending.sort_values('Amount', ascending=False)
category_spending = category_spending.reset_index(drop=True)
category_spending
```

# Task 8: Monthly profit for bank

```python
monthly_profit = pd.merge(df_spend, df_repayment, on=['Customer', 'Month'])
monthly_profit['Profit'] = monthly_profit['Repaid'] - monthly_profit['Amount']

monthly_profit['Year'] = pd.to_datetime(monthly_profit['Month']).dt.year
monthly_profit['Month'] = pd.to_datetime(monthly_profit['Month']).dt.strftime('%b')

monthly_profit = monthly_profit.groupby(['Year', 'Month'])['Profit'].sum().reset_index()
monthly_profit = monthly_profit.sort_values(['Year', 'Month'])
monthly_profit
```

# Task 9: Impose an interest rate of 2.9% for each customer for any due amount

```python
customer_due = df_spend.groupby('Customer')['Amount'].sum() - df_repayment.groupby('Customer')['Repaid'].sum()
customer_due.loc[customer_due > 0] *= 1.029
customer_due = customer_due.to_frame(name='Due_with_interest')
# Filter out negative values
customer_due = customer_due[customer_due['Due_with_interest'] > 0]
# Reset index and format the table
customer_due = customer_due.reset_index()
customer_due = customer_due.rename(columns={'Customer': 'Customer ID'})
customer_due
```
